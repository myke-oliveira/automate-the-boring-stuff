# Lendo Arquivos ZIP

Para ler o conteúdo de um arquivo ZIP, primeiro você deve criar um objeto `ZipFile` (repare nas letras mainúsculas Z e F). Objetos `ZipFile` são conceitualmente similares aos objetos `File` que você viu retornados pela função `open()` no capítulo anterior: Eles são valores através dos quais o programa interage com o arquivo. Para cirar um objeto `ZipFile`, chame a função `zipfile.ZipFile()` passando uma string de um nome de arquivo _.zip_. Repare que `zipfile` é o nome do módulo Python e `ZipFile()` é o nome da função.

```python
import zipfile, os
exampleZip = zipfile.ZipFile('Automate_the_Boring_Stuff_onlinematerials.zip')
exampleZip.namelist()
```
